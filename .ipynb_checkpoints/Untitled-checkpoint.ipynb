```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session, join
from sqlalchemy import create_engine, func
import pandas as pd
import datetime as dt
```

```python
database_path = "/Users/jonathanrocha/Desktop/SQL Alchemy/Resources/hawaii.sqlite"
engine = create_engine(f"sqlite:///{database_path}")
```

```python
base=automap_base()
```

```python
base.prepare(engine,reflect=True)
```

```python

base.metadata.tables
```

```python
conn=engine.connect()
```

```python
measurement=pd.read_sql('SELECT* FROM measurement',conn)
measurement.head()
```

```python
station=pd.read_sql('SELECT* FROM station',conn)
station.head()
```

```python
data=pd.merge(measurement,station, how='outer', on='station')
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
data.head()
```

```python
include = data[data['date'].dt.year ==2017]
include.head()
```

```python
prcp=include[['date','prcp']]
```
